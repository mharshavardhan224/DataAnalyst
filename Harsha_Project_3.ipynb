# 3. Exploratory Data Analysis (EDA)

1.	Plot the distribution of the following discrete features:
o	quality
o	Category
o	gender


```python
# Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Assuming 'cleaned_customer_shopping_data.csv' is the cleaned dataset file
data = pd.read_csv('/content/customer_shopping_data.csv')  # Load the dataset

# Plot distributions for discrete features
for col in ['gender', 'category']:
    plt.figure(figsize=(8, 6))
    sns.countplot(x=col, data=data)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()
```

2.Identify the most frequently purchased product category.

```python
# prompt: generate a code to Identify the most frequently purchased product category.

# Identify the most frequently purchased product category.
most_frequent_category = data['category'].mode()[0]
print(f"The most frequently purchased product category is: {most_frequent_category}")
```

Find the region contributing the most to sales.

```python
# prompt: GENERATE A CODE TO FIND THE REGION CONTRIBUTING THE MOST TO SALES

# Find the region contributing the most to sales.
region_sales = data.groupby('location')['purchase_amount'].sum()
highest_sales_region = region_sales.idxmax()
print(f"The region with the highest total sales is: {highest_sales_region}")
```
