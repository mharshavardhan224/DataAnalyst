```python
# prompt: generate code for drop down dash from the data set

import dash
from dash import dcc,html,Input,Output
import plotly.express as px
import pandas as pd

# Assuming you have already installed dash: pip install dash
# data = pd.read_csv('/content/customer_shopping_data.csv') # This line is already in your provided code
# data.head() # This line is already in your provided code

app = dash.Dash(__name__)

app.layout = html.Div([
    html.H1("Customer Shopping Data Analysis"),
    dcc.Dropdown(
        id='customer-dropdown',
        options=[{'label': i, 'value': i} for i in data['customer_id'].unique()],
        value=data['customer_id'].unique()[0]  # Set the initial value
    ),
    dcc.Graph(id='shopping-graph')
])

@app.callback(
    Output('shopping-graph', 'figure'),
    Input('customer-dropdown', 'value')
)
def update_graph(selected_customer):
    filtered_data = data[data['customer_id'] == selected_customer]
    fig = px.bar(filtered_data, x='invoice_date', y='price', title=f'Shopping History for Customer {selected_customer}')
    return fig

if __name__ == '__main__':
    app.run_server(debug=True)
```
